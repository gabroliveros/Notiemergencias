```python
import pandas as pd
import os
pd.set_option('display.max_rows', None, 'display.max_columns', None)
```

```python
d1 = pd.read_csv('ddg_noticias_vzla_OPEN_part1.csv')
d2 = pd.read_csv('ddg_noticias_vzla_OPEN_part2.csv', on_bad_lines='skip')
d3 = pd.read_csv('ddg_noticias_vzla_OPEN_part3.csv', on_bad_lines='skip')
d4 = pd.read_csv('ddg_noticias_vzla_OPEN_part4.csv', on_bad_lines='skip')

print('d1:', d1.shape, d1.columns.tolist())
print('d2:', d2.shape)
print('d3:', d3.shape)
print('d4:', d4.shape)
```

```python
df1 = pd.concat([d1, d2, d3, d4])
print('df1:', df1.shape)

if 'url' not in df1.columns:
    # Para Redes Sociales
    df1 = df1.drop(columns=['profile_url'], errors='ignore')
    df1[['likes', 'comments']] = ''
    df1 = df1.drop_duplicates('source_url').reset_index(drop=True)
else:
    # Para DuckDuckGo
    df1 = df1.drop_duplicates('url').reset_index(drop=True)    

print('df1:', df1.shape, df1.columns.tolist())
```

```python
df1 = df1.drop(columns=['_collected_at;;'], errors='ignore')
# df1 = df1[['source_url', 'post_url', 'username', 'profile_name', 'text', 'date', 'likes', 'comments', '_collected_at']]
print('df1:', df1.shape, df1.columns.tolist())
```

```python
# Crear el archivo contenedor y si tiene registros imprime las columnas
archivo = 'noticias_vzla_lluvias.xlsx'

if os.path.exists(archivo):
    df2 = pd.read_excel(archivo)
    print('df2:', df2.shape, df2.columns.tolist())
else:
    # Para DuckDuckGo
    # df2 = pd.DataFrame(columns=['page_number', 'rank', 'title', 'url', 'displayed_url', 'snippet', 'snippet_full', 'rich_type', 'criterio_busqueda'])
    
    # Para Redes Sociales
    df2 = pd.DataFrame(columns=["source_url", "post_url", "profile_name", "username", "text", "date", "likes", "comments", "_collected_at"])
    
    df2.to_excel(archivo, index=False)
    print(f"Archivo creado: {archivo}")
```

```python
dg = pd.concat([df2, df1])
print('dg:', dg.shape)
if 'url' in dg.columns:
    dg = dg.drop_duplicates('url').reset_index(drop=True)
else:
    dg = dg.drop_duplicates('source_url').reset_index(drop=True)
print('dg:', dg.shape)
```

```python
dg.to_excel(archivo, index=False)
```
